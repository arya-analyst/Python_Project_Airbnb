***Plotting the listings based on their price***

```python
from PIL import Image
```

```python
def eda_map(df, feature, ax):
    # Set up the figure size for the plot
    plt.figure(figsize=(15,15))
    
    # Load the background image for New York City
    image = Image.open("C:/Users/ARYA GUPTA/Desktop/Projects Data/archive (1)/New_York_City_.png")
    
    # Plot the scatter plot on the given axis
    df.plot(
        kind='scatter', 
        x='longitude', 
        y='latitude', 
        label='AirBnB On service', 
        c=str(feature),  # Color points based on the feature column
        ax=ax,           # Use the provided axis for plotting
        cmap=plt.get_cmap('jet'),  # Use 'jet' colormap
        colorbar=True,   # Show color bar
        alpha=0.4,       # Transparency of the points
        zorder=5,        # Ensure points are above the image
        s=1              # Size of the points (1 for small dots)
    )
    
    # Overlay the image on the plot
    ax.imshow(image, extent=[-74.258, -73.7, 40.49, 40.92])
```

```python
plt.figure(figsize=(20,15))
ax = plt.gca()
eda_map(df,'price',ax)
```

```python
# Initialize price range counter
price_range = 0

# Create a figure with 5 rows and 2 columns of subplots
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(18, 22))

# Loop through each subplot
for row in axes:
    for ax in row:
        if price_range < 1000:  # Only continue if price range is below $100,000
            # Filter DataFrame for listings within the current price range
            filtered_df = df[(df['price'] > price_range*100) & (df['price'] < (price_range+1)*100)]
            
            # Call the custom mapping function for this price range
            eda_map(filtered_df, 'price', ax)
            
            # Set title for subplot showing price range and count of listings
            ax.set_title(f"Price Range ${price_range*100} - ${(price_range+1)*100}, Total Listings: {len(filtered_df)}")
            
            # Increment price range for next iteration
            price_range += 1

# Adjust layout to prevent overlapping
plt.tight_layout()
```

**Checking lodging affordability**

```python
#Brooklyn
sub_1 = df.loc[(df['neighbourhood_group'] == 'Brooklyn') & (df['room_type'] == 'Private room')]
price_sub_1 = df['price'].iloc[sub_1.index] #Prices for Brooklyn neighbourhood_group

#Manhattan
sub_2 = df.loc[(df['neighbourhood_group'] == 'Manhattan') & (df['room_type'] == 'Private room')]
price_sub_2 = df['price'].iloc[sub_2.index] #Prices for Mannhattan neighbourhood_group

#Queens
sub_3 = df.loc[(df['neighbourhood_group'] == 'Queens') & (df['room_type'] == 'Private room')]
price_sub_3 = df['price'].iloc[sub_3.index] #Prices for Queens neighbourhood_group

#Staten Island
sub_4 = df.loc[(df['neighbourhood_group'] == 'Staten Island') & (df['room_type'] == 'Private room')]
price_sub_4 = df['price'].iloc[sub_4.index] #Prices for Staten Island neighbourhood_group

#Bronx
sub_5 = df.loc[(df['neighbourhood_group'] == 'Bronx') & (df['room_type'] == 'Private room')]
price_sub_5 = df['price'].iloc[sub_5.index] #Prices for Bronx neighbourhood_group


percentile_price_brooklyn = []  #percentile of prices for Private room in neighbourhood_group Brooklyn
percentile_price_manhattan = [] 
percentile_price_queens = []
percentile_price_staten_island = []
percentile_price_bronx = []


percentiles = [] 

for i in range (25,91):
    percentile_price_brooklyn.append(int(price_sub_1.quantile(i/100)))
    percentile_price_manhattan.append(int(price_sub_2.quantile(i/100)))
    percentile_price_queens.append(int(price_sub_3.quantile(i/100)))
    percentile_price_staten_island.append(int(price_sub_4.quantile(i/100)))
    percentile_price_bronx.append(int(price_sub_5.quantile(i/100)))
    percentiles.append(i)
```

```python
plt.title('Prices of 90% Private Rooms', fontsize = 12, color = 'Blue')
sns.set_style('darkgrid')

sd = sns.lineplot(x = pd.Series(percentiles), y=pd.Series(percentile_price_brooklyn), label = 'Brooklyn' )
sd = sns.lineplot(x = pd.Series(percentiles), y=pd.Series(percentile_price_manhattan), label = 'Manhattan' )
sd = sns.lineplot(x = pd.Series(percentiles), y=pd.Series(percentile_price_queens), label = 'Queens' )
sd = sns.lineplot(x = pd.Series(percentiles), y=pd.Series(percentile_price_staten_island), label = 'Staten Island' )
sd = sns.lineplot(x = pd.Series(percentiles), y=pd.Series(percentile_price_bronx), label = 'Bronx' )

sd.set(xlabel='Percentiles', ylabel='Percentile prices in $')
```

**Listing based on price-range**

```python
df['price_group'] = 0
```

```python
for idx, pri in enumerate(df['price']):
    if df.loc[idx, 'price'] < 101:
        df.loc[idx, 'price_group'] = 'Less Than 100'
    elif df.loc[idx, 'price'] < 251:
        df.loc[idx, 'price_group'] = 'bw 100 and 250'
    elif df.loc[idx, 'price'] < 501:
        df.loc[idx, 'price_group'] = 'bw 251 and 500'
    elif df.loc[idx, 'price'] < 751:
        df.loc[idx, 'price_group'] = 'bw 501 and 750'
    elif df.loc[idx, 'price'] < 1001:
        df.loc[idx, 'price_group'] = 'bw 751 and 1000'
    else:
        df.loc[idx, 'price_group'] = 'Greater Than 1000'
```

```python
x = df['price_group'].value_counts().index
y = df['price_group'].value_counts().values
plt.figure(figsize=(20,8))
for i in range(len(x)):
    height = y[i]
    plt.text(x[i], height + 0.5, '%.1f' %height, ha='center', va='bottom', size = 12)
plt.bar(x,y)
```
